```python
# 남성향 노벨수집
genres = ['urban', 'eastern', 'games', 'fantasy', 'scifi', 'horror',
         'sports', 'action', 'war', 'realistic', 'history']

df_webnovel = pd.DataFrame()

for genre in genres:
    url = "https://www.webnovel.com/stories/novel-{0}-male".format(genre)
    driver.get(url)
    driver.implicitly_wait(5) # 10초 대기
    # 스크롤 실행 전 높이
    before_h = driver.execute_script('return window.scrollY')
    before_h
    
    # 유한 스크롤
    i = 0  
    # body태그 객체를 찾고 키보드의 end키를 전송
    while i < 10 :
        # 현재 화면의 맨 아래로 스크롤을 내리기
        driver.find_element(By.CSS_SELECTOR,'body').send_keys(Keys.END)
        # 스크롤 후 페이지 로딩시간 대기
        time.sleep(2)
        i += 1 
    # 스크롤 후 페이지 로딩시간 대기
    time.sleep(2)
    # 웹소설 정보 ( 장르는 : Urban ) 장르설정은 맨 처음 url로 지정해줄 것!
    items = driver.find_elements(By.CSS_SELECTOR,"body > div.page > div.mb40 > div > div.f1 > div.j_category_wrapper > ul > li > div")
    
    title = []
    summary = []
    for item in items:
        if len(item.text.split('\n')) == 6 :
            title.append(item.text.split('\n')[1])
            summary.append(item.text.split('\n')[2])
        else :
            title.append(item.text.split('\n')[0])
            summary.append(item.text.split('\n')[1])
        
    df_webnovel = pd.concat([df_webnovel, pd.DataFrame({'genre':genre, 'title':title, 'summary':summary}, index=range(len(title)))])
```
